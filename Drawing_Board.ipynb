Program ini adalah program untuk membuat grafik berdasarkan koordinat yang diinput. 
Idenya adalah program akan menampilkan grafik kosong dengan beberapa opsi yang dapat dipilih. Selanjutnya program akan meminta pilihan dari opsi terseut secara terus menerus. Berikut opsi pilihannya.
1. Pilihan 1, program akan meminta koordinat-koordinat sumbu x dan sumbu y yang harus kita input.
2. Pilihan 2, coretan dari input sumbu x dan y akan dimunculkan dalam bentuk grafik.
3. Pilihan 3, seluruh coretan yang telah diinput akan dihapus.
4. Pilihan 4, maka program akan berhenti.

Algoritmanya:
1. Membuat suatu grafik kosong dengan sumbu x dan sumbu y : dari -4 sampai 4.
2. Membuat list kosong x dan list kosong y yang akan diisi koordinat yang diinput.
3. Menggunakan kondisi bersyarat if dari pilihan angka yang telah diinput.
4. Jika angka yang dipilih 1, input koordinat sumbu x, dan koordinat sumbu y.
5. Dengan syntax .append() koordinat yang diinput akan dimasukan ke dalam list kososng.
6. Jika angka yang dipilih 2, tampilkan plot dari koordinat x dan y.
7. Jika angka yang dipilih 3, hapus elemen dalam list dengan mengubah list menjadi list kososng kembali.
8. Jika angka yang dipilih 4, keluar dari looping untuk menghentikan program.

```python
import matplotlib.pyplot as plt #menggunakan module matplotlib.pyplot untuk membuat grafik

print('Selamat datang di program coret-coret')
print('Ini ada papan coret-coret masih kosong')
plt.xlim(-4,4) #sumbu x dari -4 sampai 4
plt.ylim(-4,4) #sumbu y dari -4 sampai 4
plt.show() #menampilkan grafik
print(' ')
print('Isi dengan titik koordinat ya')
print('1. Untuk tambahkan titik')
print('2. Untuk lihat hasil coretan')
print('3. Untuk hapus semua coretan')
print('4. Untuk menghentikan program')
print(' ')
i = 0
x = [] #list kosong untuk koordinat sumbu x
y = [] #list kosong untuk koordinat sumbu y
for i in range(1,9999999999999): #for loop dari 1 sampai tak hingga, permainan dapat terus berjalan sebelum memilih pilihan 4.
    pilihan = int(input("Pilih: ")) #untuk input pilihan
    if pilihan == 1: #kondisi bersyarat if pilihan 1
        w = eval(input('Masukkan titik-titik x: ')) #menginput list dan merubahnya menjadi kode koordinat.
        z = eval(input('Masukkan titik-titik y: '))
        x.append(w) #menambahkan elemen berupa list yang diinput pada variabel w ke dalam list kosong x
        y.append(z) #menambahkan elemen berupa list yang diinput pada variabel z ke dalam list kosong y
        print(' ')
    elif pilihan == 2: #kondisi bersyarat if pilihan = 2
        if len(x) == 0: #jika list x kosong atau panjang dari list x = 0
            plt.xlim(-4,4)
            plt.ylim(-4,4)
            plt.show()
        else: #jika list tidak kosong
            for i in range(len(x)):
                plt.plot(x[i],y[i]) #membentuk list x dan list y ke dalam plot
            plt.show()
    elif pilihan == 3: #kondisi bersyarat if pilihan 3
        x = [] #mengosongkan list x yang telah diinput elemen
        y = [] #mengosongkan list y yang telah diinput elemen
        print(' ')
    elif pilihan == 4: #kondisi bersyarat if pilihan 4
        print('Terimakasih telah bermain!')
        break #keluar dari loop supaya tidak terjadi forever loop
```
